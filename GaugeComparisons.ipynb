### Compare 2-3 Variables

```python
#X AXIS 
dataX= 'clim'
columnX='p_seasonality'

#Y AXIS
dataY='clim'
columnY='frac_snow'

#COLORS              fdc_infoALL['err_1b']
dataZ='clim'
columnZ='aridity'
colors = 4
# 0: No color code      # 3: Other, divergent scale
# 1: Dry vs not Dry     # 4: Other, linear scale
# 2: Pct Dry           

#EXCLUDE THINGS?
acceptable = subset # [] all things good

#HIGHLIGHT CHOSEN GAUGE
gauge=[] #[8025500] 

title='Primary Climate Indices'
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
fig2=comparisonplot(dataX,columnX,dataY,columnY,dataZ,columnZ,gauge,colors,s=15,acceptable=acceptable,linewidths=0.25,show=0,alpha=0.6)
plt.title(title)
plt.show()
```

### Subset/Gauge Identification

```python
#set ranges in x, y, and z to select by 
xrange=[-0.6,0.1] 
yrange=[0.4,0.8] 
zrange=[0.8,3] 
showthese =['y'] #['x','y','z'] return attribute table in thie category

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
fig=comparisonplot(dataX,columnX,dataY,columnY,dataZ,columnZ,gauge,colors,s=15,acceptable=acceptable,linewidths=0.25,show=0,alpha=0.6)
plotsquare(xrange,yrange)
plt.show()

fig=subsetID(dataX,columnX,xrange,dataY,columnY,yrange,dataZ,columnZ,zrange,colors,showthese,primaryfilter=subset)
```

### Category Designation

```python
#Designate up to 3 categories/labels

s1_label='subset1'
s1_xrange=xrange
s1_yrange=yrange
s1_marker='r--'

s2=1
if s2==1:
    s2_label='subset2'
    s2_xrange=[-1.6,-0.6]
    s2_yrange=[0,1] #[-0.25,-1]
    s2_marker='b--'
    
s3=0
if s3==1:
    s3_label='subset3'
    s3_xrange=s1_xrange
    s3_yrange=[-0.25,0.25]
    s3_marker='g--'
    
#Additional filter
bonusfilter = set(subset) # Here keeping up the original restriction of flow >98%
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
subset1=gaugefilter(dataX,columnX,s1_xrange) & gaugefilter(dataY,columnY,s1_yrange) & bonusfilter
if s2==1: 
    subset2=gaugefilter(dataX,columnX,s2_xrange) & gaugefilter(dataY,columnY,s2_yrange) & bonusfilter
if s3==1: 
    subset3=gaugefilter(dataX,columnX,s3_xrange) & gaugefilter(dataY,columnY,s3_yrange) & bonusfilter
#Note: one could instead define a subset as an arbitrary list of gauges here instead of by attribute filters if desired

#subset2.remove(8025500) #if needing to remove any gauges manually

fig=comparisonplot(dataX,columnX,dataY,columnY,dataZ,columnZ,gauge,colors,s=15,acceptable=acceptable,linewidths=0.25,show=0,alpha=0.6)

plotsquare(s1_xrange,s1_yrange,s1_marker)
if s2==1:
    plotsquare(s2_xrange,s2_yrange,s2_marker)
    plt.legend(labels=[s1_label,s2_label],bbox_to_anchor=(1.7, 0.2))
if s3==1:
    plotsquare(s3_xrange,s3_yrange,s3_marker)
    plt.legend(labels=[s1_label,s2_label,s3_label],bbox_to_anchor=(1.7, 0.2))

plt.show()
```
